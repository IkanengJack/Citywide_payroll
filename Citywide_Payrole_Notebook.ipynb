```python
import pandas as pd

# Define a function to check if a string is a valid date
def is_valid_date(value):
    try:
        pd.to_datetime(value, format="%Y-%m-%d")
        return True
    except:
        try:
            pd.to_datetime(value, format="%m/%d/%Y")
            return True
        except:
            return False

# Step 1: Identify rows where the `Work_Location` column contains dates
mask = df_cleaned['Work_Location'].apply(is_valid_date)

# Step 2: Fix misaligned rows
# Move the date from `Work_Location` to `Hire_Date`
df_cleaned.loc[mask, 'Hire_Date'] = df_cleaned.loc[mask, 'Work_Location']

# Move `Work_Location` back to its correct position from `Title_Description`
df_cleaned.loc[mask, 'Work_Location'] = df_cleaned.loc[mask, 'Title_Description']

# Clear the misaligned values in `Title_Description`
df_cleaned.loc[mask, 'Title_Description'] = None

# Step 3: Ensure `Hire_Date` is in a consistent datetime format
df_cleaned['Hire_Date'] = pd.to_datetime(df_cleaned['Hire_Date'], errors='coerce')

# Step 4: Replace missing or invalid values with defaults
df_cleaned['Work_Location'].fillna("Unknown", inplace=True)
df_cleaned['Title_Description'].fillna("Unknown", inplace=True)





```

```python
import pandas as pd

# Define functions to identify valid dates and valid leave statuses
def is_valid_date(value):
    try:
        pd.to_datetime(value, format="%Y-%m-%d")
        return True
    except:
        try:
            pd.to_datetime(value, format="%m/%d/%Y")
            return True
        except:
            return False

# Define the valid leave statuses
valid_leave_statuses = ["ACTIVE", "CEASED", "TERMINATED", "ON LEAVE", "RETIRED"]

def is_valid_leave_status(value):
    return str(value).strip().upper() in valid_leave_statuses

# Step 1: Identify rows where `Work_Location` contains dates
date_mask = df_cleaned['Work_Location'].apply(is_valid_date)

# Step 2: Fix misaligned rows for `Hire_Date`, `Work_Location`, and `Title_Description`
df_cleaned.loc[date_mask, 'Hire_Date'] = df_cleaned.loc[date_mask, 'Work_Location']
df_cleaned.loc[date_mask, 'Work_Location'] = df_cleaned.loc[date_mask, 'Title_Description']
df_cleaned.loc[date_mask, 'Title_Description'] = df_cleaned.loc[date_mask, 'Leave_Status_As_of_June_30']
df_cleaned.loc[date_mask, 'Leave_Status_As_of_June_30'] = None

# Step 3: Identify rows where `Leave_Status_As_of_June_30` is invalid
leave_status_mask = ~df_cleaned['Leave_Status_As_of_June_30'].apply(is_valid_leave_status)

# Move misaligned `Leave_Status_As_of_June_30` values back under `Title_Description`
df_cleaned.loc[leave_status_mask, 'Title_Description'] = (
    df_cleaned.loc[leave_status_mask, 'Title_Description'].fillna("") + 
    " " + 
    df_cleaned.loc[leave_status_mask, 'Leave_Status_As_of_June_30'].fillna("")
).str.strip()

# Clear the misaligned values in `Leave_Status_As_of_June_30`
df_cleaned.loc[leave_status_mask, 'Leave_Status_As_of_June_30'] = None

# Step 4: Format columns to maintain consistency
# Ensure `Hire_Date` is in a consistent datetime format
df_cleaned['Hire_Date'] = pd.to_datetime(df_cleaned['Hire_Date'], errors='coerce')

# Replace missing or invalid values with defaults
df_cleaned['Work_Location'].fillna("Unknown", inplace=True)
df_cleaned['Title_Description'].fillna("Unknown", inplace=True)
df_cleaned['Leave_Status_As_of_June_30'].fillna("Unknown", inplace=True)
```

```python
import pandas as pd
import numpy as np

# Step 1: Trim leading and trailing spaces/quotes from all string columns
df_cleaned = df_cleaned.applymap(lambda x: x.strip(' "\'`') if isinstance(x, str) else x)

# Step 2: Normalize date formats for the "Hire_Date" column
def normalize_date(date):
    try:
        # Attempt to parse the date in multiple formats
        return pd.to_datetime(date, format="%Y-%m-%d", errors="coerce") or \
               pd.to_datetime(date, format="%m/%d/%Y", errors="coerce")
    except:
        return np.nan

if "Hire_Date" in df_cleaned.columns:
    df_cleaned["Hire_Date"] = df_cleaned["Hire_Date"].apply(normalize_date)

# Step 3: Validate and clean the "Middle_Initial" column
if "Middle_Initial" in df_cleaned.columns:
    df_cleaned["Middle_Initial"] = df_cleaned["Middle_Initial"].apply(
        lambda x: x if isinstance(x, str) and len(x) == 1 and x.isalpha() else "Unknown"
    )

# Step 4: Remove special characters from specific string columns
string_columns = ["Last_Name", "First_Name", "Work_Location", "Title_Description"]
for col in string_columns:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].str.replace(r'[^a-zA-Z0-9\s\-]', '', regex=True)

# Step 5: Replace empty strings with NaN
df_cleaned.replace("", np.nan, inplace=True)

# Step 6: Fill missing values with default replacements
if "Middle_Initial" in df_cleaned.columns:
    df_cleaned["Middle_Initial"].fillna("Unknown", inplace=True)
if "Base_Salary" in df_cleaned.columns:
    df_cleaned["Base_Salary"] = pd.to_numeric(df_cleaned["Base_Salary"], errors='coerce').fillna(0.0)
if "Regular_Hours" in df_cleaned.columns:
    df_cleaned["Regular_Hours"] = pd.to_numeric(df_cleaned["Regular_Hours"], errors='coerce').fillna(0.0)

# Step 7: Remove rows where all columns are null
df_cleaned = df_cleaned.dropna(how='all')



# Preview the first few rows to confirm the cleaning process
print(df_cleaned.head())
```

```python
def fix_pay_basis_and_fill_missing(df, pay_basis_column="Pay_Basis", base_salary_column="Base_Salary"):
    # Define valid pay basis values
    valid_pay_basis = ["per Annum", "per Hour", "Prorated Annual", "per Day"]

    # Function to check if a value is a valid pay basis
    def is_valid_pay_basis(value):
        return str(value).strip() in valid_pay_basis

    # Identify rows where `Pay_Basis` contains non-standard or missing values
    pay_basis_mask = ~df[pay_basis_column].apply(is_valid_pay_basis)

    # Handle misaligned values
    df.loc[pay_basis_mask, 'Temp_Base_Salary'] = (
        df.loc[pay_basis_mask, base_salary_column]
        .fillna(0.0)  # Fill missing numeric values with float 0.0
        .astype(str) +  # Convert numeric values to strings for safe concatenation
        " " + 
        df.loc[pay_basis_mask, pay_basis_column]
        .fillna("")  # Fill missing string values with empty string
        .astype(str)
    ).str.strip()

    # Clear the misaligned values in `Pay_Basis`
    df.loc[pay_basis_mask, pay_basis_column] = None

    # Ensure `Base_Salary` is numeric and fill missing values with 0.0 (float)
    df[base_salary_column] = pd.to_numeric(df[base_salary_column], errors="coerce").fillna(0.0)

    # Fill missing values in `Pay_Basis` with a placeholder string (e.g., "Unknown")
    df[pay_basis_column].fillna("Unknown", inplace=True)

    # Drop the temporary column after fixing
    if 'Temp_Base_Salary' in df.columns:
        df.drop(columns=['Temp_Base_Salary'], inplace=True)

    return df

# Apply the function to your dataframe
df_cleaned = fix_pay_basis_and_fill_missing(df_cleaned)
```
